```python
from databaker.framework import *
```

```python
# put your input-output files here
inputfile = "annualfdi2014inwarddirectionalv3_tcm77-426424.xls"
outputfile = "example1.csv"
previewfile = "preview.html"


# current, I'm using
tabs = loadxlstabs(inputfile)
tabs = [x for x in tabs if x.name in ['1.1', '1.2', '1.3']]
for tab in tabs:
    print (tab.name)
    
```

```python
# recipe area
conversionsegments = [ ]

for tab in tabs:
    
    anchor = tab.excel_ref('D8')
    
    obs = anchor.expand(DOWN).expand(RIGHT).is_not_blank().is_not_whitespace()
    unwanted = tab.filter(contains_string ('Source: Office')) | tab.excel_ref('A10').expand(DOWN).filter(contains_string ('1')).expand(RIGHT).expand(DOWN)
    obs = obs-unwanted
    
    dimensions = [
        HDim(tab.excel_ref('B4').expand(RIGHT).is_not_blank(), TIME, DIRECTLY, ABOVE),
        HDimConst("Holding", "Value")
                ]
    
    conversionsegment = ConversionSegment(tab, dimensions, obs)
    conversionsegments.append(conversionsegment)
    
```

```python
conversionsegment = conversionsegments[0]

hdim = conversionsegment.dimensions[0]
hdim.hbagset
```

```python
for ob in list(conversionsegment.segment):
    print(conversionsegment.lookupobs(ob))
```
